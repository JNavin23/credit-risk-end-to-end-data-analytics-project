Cleaning Dataset

```python
#fixing income
df['income'] = pd.to_numeric(df['income'], errors='coerce')
```

```python
#fixing credit score
df['credit_score'] = pd.to_numeric(df['credit_score'], errors='coerce')
```

```python
#fixing date
df['last_payment_date'] = pd.to_datetime(df['last_payment_date'], errors='coerce')
```

```python
#fixing mising values
df.loc[:,'age'] = df['age'].fillna(df['age'].median())
df.loc[:,'loan_amount'] = df['loan_amount'].fillna(df['loan_amount'].median())
df.loc[:,'income'] = df['income'].fillna(df['income'].median())
df.loc[:,'credit_score'] = df['credit_score'].fillna(df['credit_score'].median())
```

```python
#fixing categorical column
df['employment_status'] = df['employment_status'].fillna('Unkbown')
```

Feature Engineering

```python
df['risk_category'] = df['credit_score'].apply(lambda x: "High Risk" if x < 650 else("Medium Risk" if x < 700 else "Low Risk"))
```

```python
today = pd.to_datetime('today')
df['days_since_last_payment'] = (today - df['last_payment_date']).dt.days
```

```python
df['income_group'] = pd.cut(df['income'],
                            bins=[0,40000,80000,120000,200000],
                            labels=['Low','Low-Mid','Upper-Mid','High'])
```
